## Pichart
- dividing each piece of data(rows,cols)into seperate and visualize

```python
branchescount=[500,600,850,250,350]
branches=['ECE','CIVIL','CSE','MECH','EEE']
colors=['r','c','b','m','g']
plt.pie(branchescount,labels=branches,startangle=90,shadow=True,explode=(0,0,0.5,0,0))
plt.show()
```

```python
branchescount=[500,600,850,250,350]
branches=['ECE','CIVIL','CSE','MECH','EEE']
colors=['r','c','b','m','g']
plt.pie(branchescount,labels=branches,startangle=90,shadow=True,explode=(0,0,0,0.3,0))
plt.show()
```

```python
branchescount=[500,600,850,250,350]
branches=['ECE','CIVIL','CSE','MECH','EEE']
colors=['r','c','b','m','g']
plt.pie(branchescount,labels=branches,startangle=90,shadow=True,explode=(0,0,0,0.3,0),autopct='%2.2f%%')
plt.show()
```

```python
branchescount=[500,600,850,250,350]
branches=['ECE','CIVIL','CSE','MECH','EEE']
colors=['r','c','b','m','g']
plt.pie(branchescount,labels=branches,startangle=90,shadow=True,explode=(0,0,0,0.3,0),autopct='%1.1f%%')
plt.show()
```

```python
branchescount=[500,600,850,250,350]
branches=['ECE','CIVIL','CSE','MECH','EEE']
colors=['r','c','b','m','g']
plt.pie(branchescount,labels=branches,startangle=90,shadow=True,explode=(0,0.5,0,0.3,0),autopct='%3.3f%%')
plt.show()
```

```python
branchescount=[500,600,850,250,350]
branches=['ECE','CIVIL','CSE','MECH','EEE']
colors=['r','c','b','m','g']
plt.pie(branchescount,labels=branches,startangle=90,shadow=True,explode=(0,0,0.5,0,0),autopct='%2.2f%%')
plt.show()
plt.savefig('piechart.png')
```
